```python
def display(data):
    print(f"The area is:{data}")

def get_input():
    length = input("Length:")
    width = input("Width:")
    return (length, width)

def compute_area(length, width):
    data = float(length) * float(width)
    return data

def main():
    (length, width) = get_input() #Its a anonymous function
    area = compute_area(length, width)
    display(area) 

main()
```

```python
def result_dispaly(ans): #This function is to display the result
    print(ans)

def take_input():        # This function is for taking input
    num1 = int(input("Enter number1:"))
    num2 = int(input("Enter number2:"))
    num3 = int(input("Enter number3:"))
    num4 = int(input("Enter number4:"))
    count = int(input("Enter the count of numbers: "))
    return (num1, num2, num3, num4, count)

def calculate_sum(num1, num2, num3, num4):        # This function is for calculating the sum of four numbers
    addition = num1 + num2 + num3 + num4    
    return addition

def calculate_avg(addition, count): # This function is for calculating the average of four numbers
    average = addition / count
    return average

def main(): # To call all the functions
    (num1, num2, num3, num4, count) = take_input()
    addition = calculate_sum(num1, num2, num3, num4)
    average = calculate_avg(addition, count)
    print(f"Average of four numbers {num1}, {num2}, {num3}, {num4} is: ", end="")
    result_dispaly(average)

main() 
```
